```python
#Creating discrete random variables and plotting probability distribution:

from scipy.stats import rv_discrete
import numpy as np
x = np.arange(5) # create a random variable x assuming values from 0 to 4
x
P_x = [0.1, 0.4, 0.3, 0.1, 0.1] # assigning probabilities
X = rv_discrete(name='X', values=(x, P_x))
print(X.pmf(x)) # Prints the probability P(X=x)
print(X.pmf(2)) # Prints the probability P(X=2)
print(X.cdf(x)) # Prints the cummulative probability P(X<=x)
```

```python
#The above probability distribution can be plotted using the following code:
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
ax.plot(x, X.pmf(x), 'go', ms=12, mec='g')
ax.vlines(x, 0, X.pmf(x), colors='g', lw=4)
plt.show()
```

**Binomial distribution:**

Python code for printing and plotting binomial distribution table and calculating mean and variance


```python
from scipy.stats import binom
import matplotlib.pyplot as plt
# setting the values of n and p
n = 6
p = 0.6
# defining the list of r values
r_values = list(range(n + 1))
# obtaining the mean and variance
mean, var = binom.stats(n, p)
# list of pmf values
dist = [binom.pmf(r, n, p) for r in r_values ]
# printing the table
print("r\tp(r)")
for i in range(n + 1):
 print(str(r_values[i]) + "\t" + str(dist[i]))
# printing mean and variance
print("mean = "+str(mean))
print("variance = "+str(var))
# plotting the graph
plt.bar(r_values, dist)
plt.show()
```

When success and failure are equally likely, the binomial distribution is a normal distribution. Hence, changing 
the value of p to 0.5, we obtain this graph, which is identical to a normal distribution plot :

**Poisson Distribution:**

Example 1: Probability Equal to Some Value
1. A store sells 3 apples per day on average. What is the probability that they will sell 5 apples on a given day?

```python
from scipy.stats import poisson
#calculate probability
poisson.pmf(k=5, mu=3)
```

Example 2: Probability Less than Some Value
1. A certain store sells seven footballs per day on average. What is the probability that this store sells four or less 
footballs in a given day?.

```python
from scipy.stats import poisson
#calculate probability
poisson.cdf(k=4, mu=7)
```

Example 3: Probability Greater than Some Value
1. A certain store sells 15 cans of tuna per day on average. What is the probability that this store sells more than 20 
cans of tuna in a given day

```python
from scipy.stats import poisson
#calculate probability
1-poisson.cdf(k=20, mu=15)
```

## Normal Distribution:

Generate a Normal Distribution in Python

```python
from numpy.random import normal
import numpy as np
import matplotlib.pyplot as plt
#generate sample of 200 values that follow a normal distribution
data = normal(loc=0, scale=1, size=200)
#find mean of sample
np.mean(data)
#find standard deviation of sample
np.std(data, ddof=1)
# Creating a histogram to visualize the distribution of data
plt.hist(data, 30)
plt.show()
```

Plotting a Single Normal Distribution:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#x-axis ranges from -3 and 3 with .001 steps
x = np.arange(-3, 3, 0.001)
#plot normal distribution with mean 0 and standard deviation 1
plt.plot(x, norm.pdf(x, 0, 1), color='red', linewidth=3)
```

Example 2: Plotting Multiple Normal Distributions


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#x-axis ranges from -5 and 5 with .001 steps
x = np.arange(-5, 5, 0.001)
#define multiple normal distributions
plt.plot(x, norm.pdf(x, 0, 1), label=': 0, : 1')
plt.plot(x, norm.pdf(x, 0, 1.5), label=':0, : 1.5')
plt.plot(x, norm.pdf(x, 0, 2), label=':0, : 2')
#add legend to plot
plt.legend(title='Parameters')
```
